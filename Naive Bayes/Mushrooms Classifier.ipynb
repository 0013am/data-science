```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('mushrooms.csv')
print(df.shape)
df.head()
```

Encode the Categorial Data into Numerical Data

```python
le=LabelEncoder()
ds=df.apply(le.fit_transform)
ds
```

```python
Xds=ds.iloc[:,1:].values
yds=ds.iloc[:,0].values
print(Xds.shape,yds.shape)
Xds
```

Break the Data into train and test

```python
X,x,Y,y=train_test_split(Xds,yds,test_size=0.2)
```

```python
X.shape,Y.shape,x.shape,y.shape
```

```python
np.unique(y)
```

Calculating prior and cond_prob 

```python
def prior(Y,label):
    cnt=np.sum(Y==label)
    total=Y.shape[0]
    return (cnt)/float(total)
```

```python
def cond(X,Y,f_c,f_v,label):
    x_f=X[Y==label]
    num=np.sum(x_f[:,f_c]==f_v)
    den=np.sum(Y==label)
    return (num)/float(den)
```

Predication

```python
def pred(X,Y,x):
    classes=np.unique(Y)
    n=X.shape[1]
    post=[]
    for c in classes:
        likelihood=1
        for i in range(n):
            con=cond(X,Y,i,x[i],c)
            likelihood*=con
        pri=prior(Y,c)
        ps=likelihood*pri
        post.append(ps)
    pre=np.argmax(post)
    return pre
```

```python
out=pred(X,Y,x[14])
print(out)
y[14]
```

Score

```python
def score(X,Y,y,x):
    out=[]
    for i in range(x.shape[0]):
        pre=pred(X,Y,x[i])
        out.append(pre)
    out=np.array(out)
    cnt=np.sum(out==y)
    total=y.shape[0]
    return cnt/total
```

```python
score(X,Y,y,x)
```
